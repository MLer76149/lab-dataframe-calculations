## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
orders = pd.read_csv("Orders.csv")
```

```python
orders = orders.drop(columns=["Unnamed: 0"])
```

```python
display(orders)
```

```python
customer_spent = orders.groupby(["CustomerID", "Country"]).agg({"amount_spent": sum}).reset_index()
```

```python
display(customer_spent)
```

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
quant95 = np.quantile(customer_spent["amount_spent"], .95)
quant75 = np.percentile(customer_spent["amount_spent"], 75)
```

```python
def lable(row):
    if row > quant95:
        return "VIP"
    elif row > quant75 and row < quant95:
        return "Preffered"
    else:
        return "No label"
```

```python
customer_spent["Lable"] = customer_spent["amount_spent"].apply(lable)
```

```python
customer_spent["Lable"].value_counts()
```

```python
display(customer_spent)
```

## Q2: How to identify which country has the most VIP Customers?

```python
lable_by_country = customer_spent.groupby(["Country", "Lable"]).agg({"Lable": pd.value_counts}).rename(columns={"Lable": "Lable Count"}).reset_index()
```

```python
display(lable_by_country)
```

```python
country_most_VIP = lable_by_country[(lable_by_country["Lable"] == "VIP")]
```

```python
display(country_most_VIP)
```

```python
country_most_VIP[country_most_VIP["Lable Count"]==country_most_VIP["Lable Count"].max()]
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
country_most_VIP_Pref = lable_by_country[(lable_by_country["Lable"].isin(["VIP", "Preffered"]))].groupby("Country").agg({"Lable Count": sum}).reset_index()
```

```python
display(country_most_VIP_Pref)
```
